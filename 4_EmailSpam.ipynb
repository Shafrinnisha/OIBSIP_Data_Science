```python
#Import python liabraries from scikit-learn.
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer, CountVectorizer
from sklearn.naive_bayes import MultinomialNB
```

```python
#Load dataset
df=pd.read_csv("spam.csv",encoding="latin1")
df.sample(5)
```

```python
#check descriptive statistics
df.info()
```

```python
#Lets see null value count in df
df.isnull().sum()
```

```python
df.isnull().mean()*100  #check the percentage of null value
```

```python
# Drop unnecessary columns
df.drop(columns=df[['Unnamed: 2','Unnamed: 3','Unnamed: 4']],axis=1,inplace=True)
```

```python
# Display the first few rows of the dataset before encoding
df.head(2)
```

```python
# Rename columns for better understanding
df.columns = ['Category', 'Message']

# Encode target labels
encoder = LabelEncoder()
df['Category'] = encoder.fit_transform(df['Category'])
```

```python
# Display the first few rows of the dataset after encoding
df.head(2)
```

0 is ham and 1 is spam

```python
# Split the data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(df.Message, df.Category, test_size=0.2)
```

```python
# Vectorize text data using TF-IDF
vectorizer = TfidfVectorizer()
X_train_vec = vectorizer.fit_transform(X_train)
X_test_vec = vectorizer.transform(X_test)
```

```python
# Train a Naive Bayes model
model = MultinomialNB()
model.fit(X_train_vec, y_train)
```

```python
accuracy = model.score(X_test_vec, y_test)
print(f'Accuracy of the model: {accuracy * 100:.2f}%')
```

```python
# CountVectorizer example
vec = CountVectorizer()
X_train_count = vec.fit_transform(X_train.values)
X_test_count = vec.transform(X_test)
```

```python
model.fit(X_train_count, y_train)
```

```python
accuracy = model.score(X_test_count, y_test)
print(f'Accuracy of the model: {accuracy * 100:.2f}%')
```
